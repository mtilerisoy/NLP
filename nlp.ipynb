```python
def HammingDistance(string1, string2) -> int:
    len1 = len(string1)     # length of string1

    hammDis = 0

    for i in range(len1):               # for the length of the string
        if(string1[i] != string2[i]):   # compare all the characters one by one
            hammDis = hammDis + 1

    print(hammDis)
    return hammDis
    raise NotImplementedError()
```
